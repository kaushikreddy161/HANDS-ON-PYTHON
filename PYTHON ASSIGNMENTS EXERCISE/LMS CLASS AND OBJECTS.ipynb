## 2. How to call a variable and function by creating a simple class


```python
class Dog:
    def __init__(self, variable1):
        self.variable1 = variable1  # variable

    def show(self):
        print("I'm from function")
        print("This is Variable1:", self.variable1)

# create object
d1 = Dog("Hello Dog")

# call function
d1.show()
```

## 3. Create a class with one `static variable` and two `instace variables`

```python
class Demo:
    # static variable
    static_var = "I am Static variable"

    def __init__(self, instance_var1, instance_var2):
        # instance variables
        self.instance_var1 = instance_var1
        self.instance_var2 = instance_var2

# create object
d1 = Demo("instance_var1", "instance_var2")

# access static variable
print(Demo.static_var)

# access instance variables
print(d1.instance_var1)
print(d1.instance_var2)
```

## 6. Create a class for `Students` information like `name, rollno, marks` and `school name` with static variable.
- create a function for displaying student information
- create one more function for getting students marks and percentage 

```python
class Student: 
    def __init__(self, name, rollno, marks, school_name): 
        self.name = name 
        self.rollno = rollno 
        self.marks = marks 
        self.school_name = school_name 

    # Method to calculate percentage out of 500
    def get_percentage(self):
        percentage = (self.marks / 500) * 100
        return f"{percentage:.2f}%"

s1 = Student('abc', 1, 399, 'DIT') 

# Access instance variables
print(f"Name: {s1.name}") 
print(f"Roll No: {s1.rollno}")
print(f"Marks: {s1.marks}")
print(f"School: {s1.school_name}")

# Access the included percentage method
print(f"Percentage: {s1.get_percentage()}")
```

## 8. Create a class `Employee` with two arguments `name and salary`
- create a function which displays the employee information
- Print total number of employees 

```python
class Employee:
    # Class variable to keep track of total employees
    employee_count = 0

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        # Increase the count every time a new employee object is created
        Employee.employee_count += 1

# Create the employee instances
emp1 = Employee("Zara", 2000)
emp2 = Employee("Manni", 5000)

# Print the required output
print(f"Name : {emp1.name} , Salary: {emp1.salary}")
print(f"Name : {emp2.name} , Salary: {emp2.salary}")
print(f"Total Employee {Employee.employee_count}")

   #CODE HERE
```

```python
# Find the list of letter starting with 
#- u
#- o
words = text.split()
# You must define BOTH here
u_words = set()
o_words = set() 

for word in words:
    lowered = word.lower()
    if lowered.startswith('u'):
        u_words.add(lowered)
    elif lowered.startswith('o'):
        o_words.add(lowered)

print("List of words starting with u is:")
print(u_words)
print("List of words starting with o is:")
print(o_words)
```

```python
# Find the list of letter ends with 
#- e
#- n

words = text.split()

# You must define BOTH here
e_words = set()
n_words = set() 

for word in words:
    lowered = word.lower()
    if lowered.endswith('e'):
        e_words.add(lowered)
    elif lowered.endswith('n'):
        n_words.add(lowered)

print("List of words ends with e is:")
print(e_words)
print("List of words ends with n is:")
print(n_words)
```

```python
# 1. Split into sentences and clean the '\n'
# We use a list comprehension to strip the newline characters from each sentence
sentence = [s.replace('\n', ' ').strip() for s in text.split('. ')]

# 2. Print the first word from each sentence
print("First words from each sentence:")
for s in sentence:
    if s: # Checks if the sentence is not empty
        first_word = s.split()[0]
        print(first_word)
```

```python
# 3. Print even-positioned sentences (Index 0, 2, 4...)
# In Python, index 0 is considered the 1st (even) position in many logic tasks
print("\nEven-positioned sentences:")
even_sentences = sentence[0::2] 
for s in even_sentences:
    print(f"- {s}")
```

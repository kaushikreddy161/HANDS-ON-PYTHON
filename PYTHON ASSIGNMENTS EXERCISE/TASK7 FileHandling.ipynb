<h4>create a file 'File.txt' which explains python file handling whatever you have learnt today(creating,reading,writing and appending...etc) in detail.That reload the saved file(i.e File.txt) to jupyter notebook.</h4>

```python
# Content about Python file handling
content = """
Python File Handling

1. Creating a File
Python can create files using open() with 'w' or 'x' mode.

2. Reading a File
Use 'r' mode to read file contents.

3. Writing to a File
Use 'w' mode to write data into a file.

4. Appending to a File
Use 'a' mode to add data without deleting old data.

5. File Modes
r - Read
w - Write
a - Append
x - Create

6. Using with open()
It automatically closes the file after use.
"""

# Create and write into File.txt
with open("File.txt", "w") as file:
    file.write(content)

print("File created successfully!")

# Reload and read the file
with open("File.txt", "r") as file:
    data = file.read()

print("\nContents of File.txt:\n")
print(data)
```

<h4>search and Replace Create a Python program that reads a text file named "data.txt" and asks the user to enter a word to search for. If the word is found in the file, replace all occurrences of the word with a new word provided by the user.</h4>

```python
# Read the contents of data.txt

with open("data.txt", "r") as file:
    content = file.read()

# Ask user for search word and replacement word
search_word = input("Enter the word to search: ")
replace_word = input("Enter the new word: ")

# Check if the word exists in the file
if search_word in content:
    # Replace all occurrences
    updated_content = content.replace(search_word, replace_word)

    # Write updated content back to the file
    with open("data.txt", "w") as file:
        file.write(updated_content)

    print("Word replaced successfully!\n")

    # Display updated content
    print("Updated File Content:\n")
    print(updated_content)

else:
    print("Word not found in the file.")
```

<h4>Append to a File Write a Python program that prompts the user to enter a sentence. Append this sentence to an existing file named "notes.txt". Make sure the newly added sentence starts on a new line.
</h4>

```python
# Prompt the user to enter a sentence
sentence = input("Enter a sentence: ")

# Open notes.txt in append mode and add the sentence on a new line
with open("notes.txt", "a") as file:
    file.write("\n" + sentence)

print("Sentence appended successfully!")
```

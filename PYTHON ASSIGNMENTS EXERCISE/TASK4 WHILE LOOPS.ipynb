<h5>Create a program that prompts the user for an integer input and then uses a while loop to print the countdown from that number to 1.</h5>


```python
# Prompt the user for an integer input
# We use int() to convert the input string into a number
number = int(input("Enter a number: "))

# Use a while loop to print the countdown from the number to 1
while number >= 1:
    print(number)
    # Decrement the number by 1 in each iteration
    number -= 1
```

<h5>Create a program that uses a while loop to print the numbers from 1 to 20, but skip printing the numbers 5, 10, and 15 using the continue statement.</h5>

```python
number = 1

while number <= 20:
    if number in (5, 10, 15):
        number += 1  # Increment before continuing to avoid an infinite loop
        continue     # Skip the rest of the loop and jump to the next iteration

    # Print the allowed numbers
    print(number)
    
    # Increment the counter normally
    number += 1

    #for i in range(1,21):
  #if i not in (5,10,15):
      #print(i)
```

<h3>Calculate the Sum of Numbers from 1 to N: Write a Python program that calculates and prints the sum of all numbers from 1 to N, where N is a positive integer entered by the user, using a for loop.</h3>


```python
n = int(input("Enter a positive number: "))
sum = 0

for i in range(1, n + 1):
    sum = sum + i

print(sum)
```

<h3>Print Even Numbers: Write a Python program that prints all even numbers from 50 to 100 using a for loop.</h3>

```python
for i in range(50,100):
 if i%2==0:
    print(i)
```

<h3>reverse a String: Write a Python program that takes a string as input and uses a for loop to print the characters of the string in reverse order.</h3>



```python
text = input("Enter a string: ")

for char in reversed(text):
    print(char, end="")
```

<h3>Count Vowels in a String: name = "innomatics research labs" Write a Python program that uses a for loop to count the number of vowels (a, e, i, o, u) in the string. Ignore case sensitivity.
</h3>

```python
name = "innomatics research labs"
count=0
for char in name:
    count+=1 
print(count)
```

```python
name = "innomatics research labs"
count=0
for char in name:
    if char in "aeiou":
        count+=1
print(count)
```

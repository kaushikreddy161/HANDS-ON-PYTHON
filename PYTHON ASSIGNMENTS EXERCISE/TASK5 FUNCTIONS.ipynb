<h5>Write a function called "count_vowels" that takes a string as input and returns the number of vowels (a, e, i, o, u) in that string.
</h5>

```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count


print(count_vowels("kaushik"))
```

<h5>Write a Python function to check whether a number falls within a given range</h5>

```python
def in_range(num, start, end):
    return start <= num <= end


print(in_range(5, 1, 10))


print(in_range(12, 1, 10))
```

<h5>Write a Python function that accepts a string and counts the number of upper and lower case letters</h5>

```python
def count_letters(text):
    upper_count = 0
    lower_count = 0

    for char in text:
        if char.isupper():
            upper_count += 1
        if char.islower():
            lower_count += 1

    print("Uppercase letters:", upper_count)
    print("Lowercase letters:", lower_count)


# Test the function
count_letters("Hello World!")
```
